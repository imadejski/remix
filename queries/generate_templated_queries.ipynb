```python
import pandas as pd
```

```python
labels = [
    "Atelectasis",
    "Cardiomegaly",
    "Consolidation",
    "Edema",
    "Enlarged Cardiomediastinum",
    "Fracture",
    "Lung Lesion",
    "Lung Opacity",
    "No Finding",
    "Pleural Effusion",
    "Pleural Other",
    "Pneumonia",
    "Pneumothorax",
    "Support Devices",
]
```

```python
# both biovil and gloria tokenizers do_lower_case so capitalization does not matter
templates = [
    "Findings consistent with {label}",
    "Findings suggesting {label}",
    "Findings are most compatible with {label}",
    "{label} seen",
]
```

```python
data = []
for label in labels:
    for template in templates:
        data.append((label, template.format(label=label)))
df = pd.DataFrame(data, columns=["Variable", "Text"])
```

```python
df.to_csv("templated.csv", index=False)
```

```python
pd.DataFrame.from_dict({"Variable": labels, "Text": labels}).to_csv("direct.csv", index=False)
```
