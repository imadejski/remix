```python
chexpertplus_joined = metadata.merge(split, on=["subject_id", "study_id", "dicom_id"], how="left")
chexpertplus_frontal_only = chexpertplus_joined[chexpertplus_joined["frontal_lateral"] == "Frontal"]
```

```python
report = pd.read_csv("/opt/gpudata/chexpertplus/report.csv")
```

```python
report
```

```python
has_impr = report["impression"].astype(str).str.strip().ne("") & report["impression"].notna()
total = len(report)
num_no_impr = (~has_impr).sum()
pct_no_impr = (num_no_impr / total) * 100
print(f"Total studies in report: {total}")
print(f"No impression section: {num_no_impr} ({pct_no_impr:.2f}%)")
```

```python
studies_with_impr = set(report.loc[has_impr, "study_id"].unique())
chexpert_plus_frontal_impression_only = chexpertplus_frontal_only[chexpertplus_frontal_only["study_id"].isin(studies_with_impr)].copy()
```
