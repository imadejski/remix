```python
# 2) Combine and plot (same logic as before)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import colorsys
from matplotlib.patches import Patch

frames = []
for model_name, df in model_to_df.items():
    assert {"label", "metric", "value"}.issubset(df.columns), f"{model_name} missing cols"
    tmp = df.copy()
    tmp["model"] = model_name
    frames.append(tmp)
results_df = pd.concat(frames, ignore_index=True)

def parse_metadata(model_name: str):
    name_lower = model_name.lower()
    if "biovil" in name_lower:
        family = "biovil"
    elif ("chexpert" in name_lower) or ("gloria" in name_lower):
        family = "gloria"
    else:
        family = "other"

    if "-no-mlm" in name_lower:
        mlm = "no-mlm"
    elif "-mlm" in name_lower:
        mlm = "mlm"
    else:
        mlm = "none"

    variant_key = model_name.replace("-no-mlm", "").replace("-mlm", "")
    return family, mlm, variant_key

meta = results_df["model"].apply(parse_metadata)
results_df["family"] = [m[0] for m in meta]
results_df["mlm"] = [m[1] for m in meta]
results_df["variant_key"] = [m[2] for m in meta]

# Build model_order with custom priority:
def custom_priority(model_name: str) -> tuple:
    """Return a tuple that enforces biovil -> original-ig-tgl -> v2-ig_tg -> rest."""
    name = model_name.lower()
    if name.startswith("biovil-base"):
        return (0, 0)  # top priority
    elif name.startswith("original-ig-tgl"):
        # mlm before no-mlm
        return (1, 0 if "-mlm" in name and "-no-mlm" not in name else 1)
    elif name.startswith("v2-ig_tg"):
        return (2, 0 if "-mlm" in name and "-no-mlm" not in name else 1)
    else:
        # fallback: after these three groups, use old family/variant ordering
        fam, mlm_tag, key = parse_metadata(model_name)
        base_rank = (3, family_rank(fam), align_rank(key), version_rank(key), key)
        if mlm_tag == "mlm":
            return base_rank + (0,)
        elif mlm_tag == "no-mlm":
            return base_rank + (1,)
        else:
            return base_rank + (2,)

# sort models by priority
all_models = results_df["model"].drop_duplicates().tolist()
model_order = sorted(all_models, key=custom_priority)


# --- Color setup (one base per variant_key; muted for no-MLM) ---
# Paul Tol "Bright" 7-color palette (RGB 0-1)
TOL_BRIGHT_7 = [
    (0.000, 0.447, 0.698),  # blue
    (0.835, 0.369, 0.000),  # orange
    (0.000, 0.620, 0.451),  # green
    (0.800, 0.475, 0.655),  # pink
    (0.941, 0.894, 0.259),  # yellow
    (0.337, 0.706, 0.914),  # sky
    (0.000, 0.620, 0.980),  # cyan-ish
]

# Map distinct variant_keys (your 7 groups) to palette colors in the order of variant_order
variant_keys = [vk for _, vk in variant_order]
variant_keys = list(dict.fromkeys(variant_keys))  # preserve order, drop dups

key_to_color = {
    k: TOL_BRIGHT_7[i % len(TOL_BRIGHT_7)]
    for i, k in enumerate(variant_keys)
}

def desaturate_and_lighten(rgb, desat=0.45, lighten=0.18):
    """Reduce saturation and raise lightness a bit to create a 'muted' tone."""
    r, g, b = rgb
    h, l, s = colorsys.rgb_to_hls(r, g, b)
    s = max(0.0, s * (1.0 - desat))        # reduce saturation
    l = min(1.0, l + lighten)              # raise lightness
    return colorsys.hls_to_rgb(h, l, s)

model_to_color = {}
for m in results_df["model"].drop_duplicates():
    _, mlm_tag, key = parse_metadata(m)
    base = key_to_color.get(key, (0.5, 0.5, 0.5))
    if mlm_tag == "no-mlm":
        model_to_color[m] = desaturate_and_lighten(base, desat=0.45, lighten=0.18)
    else:
        model_to_color[m] = base

metrics = [
    "average_cosine_similarity_mean_accuracy",
    "average_cosine_similarity_mean_dcg_n",
    "average_cosine_similarity_mean_accuracy_top_10",
]
label_order = sorted(results_df["label"].unique())

sns.set_theme(style="whitegrid")
for metric in metrics:
    dfm = results_df[results_df["metric"] == metric].copy()
    if dfm.empty:
        print(f"No rows for metric: {metric}")
        continue

    plt.figure(figsize=(18, 6))
    ax = sns.barplot(
        data=dfm,
        x="label",
        y="value",
        hue="model",
        order=label_order,
        hue_order=model_order,
        palette=model_to_color,   # dict: model -> color
        edgecolor="black",
        linewidth=0.2,
    )
    ax.set_title(metric)
    ax.set_xlabel("Label")
    ax.set_ylabel("Value")
    plt.xticks(rotation=45, ha="right")

    handles, labels = ax.get_legend_handles_labels()
    legend_title = "Per pair: base=MLM, muted=no-MLM"
    plt.legend(handles, labels, title=legend_title, bbox_to_anchor=(1.01, 1), loc="upper left")

    plt.tight_layout()
    plt.show()
```
